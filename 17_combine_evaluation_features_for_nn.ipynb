## Combine features (evaluation)<br>

In what follows, we collect all the evaluation features extracted in previous steps and we combine them. For this purpose, we take extra care to exclude any pair of articles that misses one of the features from our collection.

```python
import os
from bs4 import BeautifulSoup
import json
import pandas as pd
import numpy as np
import math
```

```python
full_df = pd.read_csv ('eval/_EVAL_details_in_df.csv')
incomplete_df = pd.read_csv ('eval/_EVAL_incomplete_pairs_df.csv')
# title features
title_cos_df = pd.read_csv ('eval/_EVAL_title_cosSim_score.csv')
title_ne_df = pd.read_csv ('eval/_EVAL_title_named-entity_score.csv')
# keyword feature
key_df = pd.read_csv ('eval/_EVAL_keywords_score.csv') # when one in pair misses keys, score is -1
# text feature
text_cos_df = pd.read_csv('eval/_EVAL_text_cosSim_score.csv')
text_ne_df = pd.read_csv('eval/_EVAL_text_named-entity_score.csv')
```

```python
print(len(full_df))
print(len(incomplete_df))

print(len(title_cos_df))
print(len(title_ne_df))

print(len(key_df))

print(len(text_ne_df))
print(len(text_cos_df))
```

```python
def normal_round(n):
    if n - math.floor(n) < 0.5:
        return int(math.floor(n))
    return int(math.ceil(n))

features_df = pd.DataFrame(columns = ["pair_id","overall","title_cos","title_ne","key","text_cos","text_ne"])

dfs = [title_cos_df,title_ne_df,key_df,text_ne_df,text_cos_df]

for i, row in full_df.iterrows(): 
    pair = row['pair_id']
#     # make sure they line up 
#     if len([df.iloc[i]['pair_id'] for df in dfs if df.iloc[i]['pair_id'] != pair])>0:
#         print('don\'t line up')
#         break

    # copy needed features into one datafram
    label = normal_round(row['Overall'])
    title_cos = title_cos_df.iloc[i]['cos_sim']
    title_ne = title_ne_df.iloc[i]['key_score']
    key = key_df.iloc[i]['score']
    text_cos = text_cos_df.iloc[i]['text_cos_sim']
    text_ne = text_ne_df.iloc[i]['key_score']
    print('---------------------',i,'-----------------------')
    print(pair)
    print('title_cos:',title_cos)
    print('title_ne:',title_ne)
    print('key:', key)
    print('text_cos:',text_cos)
    print('text_key:',text_ne)
    print('label:',label)
    entry = { "pair_id": pair,"overall":label,"title_cos":title_cos,"title_ne":title_ne,"key":key,"text_cos":text_cos,"text_ne":text_ne }
    features_df = features_df.append(entry, ignore_index = True)
```

```python
path = 'eval/_EVAL_FULL_features_df.csv'
features_df.to_csv(path,index=False)  
features_df
```

```python
# remove pairs that have features missing (in incomplete pairs, have nan, key==-1)

incomplete_list = incomplete_df['pair_id'].tolist()

def incomplete(row):
    
    if row['pair_id'] in incomplete_list:
        return True
    cols = ['overall','title_cos','title_ne','key','text_cos','text_ne']
    for col in cols:
        if row[col] < 0:
            return True
        if np.isnan(row[col]):
            return True
        if pd.isnull(row[col]):
            return True
    return False
       
        
features_complete_df = pd.DataFrame(columns = ["pair_id","overall","title_cos","title_ne","key","text_cos","text_ne"])

for i, row in features_df.iterrows(): 
    if incomplete(row):
        continue
    pair = row['pair_id']
    label = row['overall']
    
    entry = { "pair_id": pair,"overall":label,"title_cos":row['title_cos'],"title_ne":row['title_ne'],"key":row['key'],"text_cos":row['text_cos'],"text_ne":row['text_ne'] }
    features_complete_df = features_complete_df.append(entry, ignore_index = True)
     
```
