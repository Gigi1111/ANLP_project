```python
import pandas as pd

pair_df = pd.read_csv ('semeval-2022_task8_train-data_batch.csv')
pair_df
```

```python
path = '1_training_csv/all_training_with_translated_title.csv'
train_df = pd.read_csv(path)
train_df
```

```python
train_dict = {}
for index, row in train_df.iterrows():
    row['id'] = str(row['id'])
    train_dict[row['id']] = {"title":row["title"],"meta_lang":row["meta_lang"],"translated_lang":row["translated_lang"],
                             "translated_title":row["translated_title"]
                             }
    
```

```python
train_dict['1484067970']
```

```python
# Geography, Entities, Time,Narrative,Overall,Style,Tone
import numpy as np

pair_df['title1'] = pair_df['title1'].astype(str)
pair_df['title2'] = pair_df['title2'].astype(str)
for i, row in pair_df.iterrows():
    ids = row['pair_id'].split('_')

    pair_df.at[i,'id1'] = str(ids[0])
    pair_df.at[i,'id2'] = str(ids[1])
    
    if str(ids[0]) in train_dict.keys():
        pair_df.at[i,'title1'] = train_dict[ids[0]]['translated_title']
    else: 
        pair_df.at[i,'title1'] = np.nan
        
    if str(ids[1]) in train_dict.keys():
        pair_df.at[i,'title2'] = train_dict[ids[1]]['translated_title']
    else: 
        pair_df.at[i,'title2'] = np.nan
        
```
