```python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
import pandas as pd
```

```python
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
# get cosine similarity from Siamese BERT
sentences = [
    "Three years later, the coffin was still full of Jello.",
    "The fish dreamed of escaping the fishbowl and into the toilet where he saw his friend go.",
    "The person box was packed with jelly many dozens of months later.",
    "He found a leprechaun in his walnut shell."
]
sentence_embeddings = model.encode(sentences)
sentence_embeddings.shape

cosine_similarity(
    [sentence_embeddings[0]],
    sentence_embeddings[1:]
)
```

```python
print(len(pair_df[pair_df['title1'].isna()]))
missing_title1_df = pair_df[pair_df['title1'].isna()]
missing_title1_df
print(len(pair_df[pair_df['title2'].isna()]))
missing_title2_df = pair_df[pair_df['title2'].isna()]
missing_title2_df
```

```python
# get cosine similarity from Siamese BERT
import numpy as np
data = []

for i, row in pair_df.iterrows():
    print('--------------',i,'---------------')
    cos_sim = 0
    sentences = [
        str(pair_df.iloc[i]['title1']),
        str(pair_df.iloc[i]['title2']),
    ]
    
    if missing_title1_df['pair_id'].str.contains(row['pair_id']).any() or missing_title2_df['pair_id'].str.contains(row['pair_id']).any():
        cos_sim = -1
    else: 
        sentence_embeddings = model.encode(sentences)

        cs = cosine_similarity(
            [sentence_embeddings[0]],
            sentence_embeddings[1:]
        )
        cos_sim = cs[0][0]
    
    for s in sentences:
        print(s)
    print('narrative: ',pair_df.iloc[i]['Narrative'])
    print('overall: ',pair_df.iloc[i]['Overall'])
    print('cos_sim: ',cos_sim)
    
    data.append([pair_df.iloc[i]['id1'],pair_df.iloc[i]['id2'],pair_df.iloc[i]['title1'],pair_df.iloc[i]['title2'],pair_df.iloc[i]['Narrative'],pair_df.iloc[i]['Overall'],cs[0][0]])

df = pd.DataFrame(data, columns=['id1','id2','title1','title2','narrative','overall','cosine_similarity'])
df
```

```python
path = '1_training_csv/title_siamese_bert_cosine_similarity.csv'
df.to_csv(path,index=False)  
```
